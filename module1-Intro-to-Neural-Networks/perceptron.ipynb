```python
from typing import Union

def step(x: np.number) -> np.number: 
    if x>0: return 1
    elif x==0: return 0
    else: return -1
    
def nand(x1: Union[bool, int], x2: Union[bool, int]) -> Union[bool, int]: 
    feedthru1 = bool(x1)
    feedthru2 = bool(x2)
    AND = bool(x1) and bool(x2)
    return step(sum([int(feedthru1), -2 * int(AND), int(feedthru2)]))
    #return int(bool(x1) or bool(x2)) - 2 * int(bool(x1) and bool(x2))

for x,y in [(0,0), (0,1), (1,0), (1,1)]: 
    print(nand(x,y))
```
