## Write your own perceptron code that can correctly classify (99.0% accuracy) a NAND gate. 

| x1 | x2 | y |
|----|----|---|
| 0  | 0  | 1 |
| 1  | 0  | 1 |
| 0  | 1  | 1 |
| 1  | 1  | 0 |

```python
# Import statements
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
import random

data = { 'x1': [0,1,0,1],
         'x2': [0,0,1,1],
         'y':  [1,1,1,0]
       }

correct_outputs = [[1], [1], [1], [0]]
# correct_outputs = [[value] for value in data['y']]

df = pd.DataFrame.from_dict(data).astype('int')

df
```

```python
X = df[['x1', 'x2']]
X['bias'] = [1,1,1,1] # Simulated bias
y = df['y']

X.head()
```

```python
inputs = X[['x1', 'x2', 'bias']].values

inputs
```

```python
# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    sx = sigmoid(x)
    return sx * (1-sx)
```

### Initialize random weights

```python
weights = 2 * np.random.random((3,1)) - 1

weights
```

### Calculate weighted sum of inputs and weights

```python
weighted_sum = np.dot(inputs, weights)

weighted_sum
```

### Training first epoch

```python
activated_output = sigmoid(weighted_sum) # Prediction first epoch

activated_output
```

### Calculate error

```python
error = correct_outputs - activated_output # Absolute error

error
```

### Gradient descent (weights correction)

```python
adjustments = error * sigmoid_derivative(weighted_sum)
```

```python
weights = 0 # Clear

weights += np.dot(inputs.T, adjustments)

weights
```

### Update weights 'n' times

```python
for iteration in range(10000):
    
    # Weighted sum of inputs / weights
    weighted_sum = np.dot(inputs, weights)
    
    # Activate!
    activated_output = sigmoid(weighted_sum)
    
    # Calculate absolute error
    error = correct_outputs - activated_output
    
    adjustments = error * sigmoid_derivative(weighted_sum)
    
    # Update the Weights
    weights += np.dot(inputs.T, adjustments)
    
print("Weights after training")
print(weights)

print("Output after training")
print(activated_output)
```
